```python
# Databricks notebook source

print("Creating Employee Table...")

spark.sql("""
CREATE TABLE IF NOT EXISTS default.employee (
    EmpId STRING,
    EmpName STRING,
    Department STRING,
    Salary STRING
)
USING DELTA
""")

print("Employee Table Created.")

print("Creating Metadata Table...")

spark.sql("""
CREATE TABLE IF NOT EXISTS default.metadata (
    TableName STRING,
    SourceFile STRING,
    RowsLoaded BIGINT,
    Status STRING,
    LoadDate STRING
)
USING DELTA
""")

print("Metadata Table Created.")

print("All Tables Created Successfully.")
```
