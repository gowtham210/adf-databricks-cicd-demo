```python
df = spark.read.format("csv") \
    .option("header", "true") \
    .load("abfss://data@stgowthamdev01.dfs.core.windows.net/employee.csv")
display(df)
```

```python
df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("employee")
```

```python
spark.sql("SELECT * FROM employee").display()
```

```python
from pyspark.sql import Row
from datetime import datetime

metadata = [
    Row(
        TableName="employee",
        SourceFile="employee.csv",
        RowsLoaded=df.count(),
        Status="SUCCESS",
        LoadDate=str(datetime.now())
    )
]

metadata_df = spark.createDataFrame(metadata)

metadata_df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("metadata")
```
