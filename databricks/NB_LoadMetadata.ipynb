```python
# Databricks notebook source

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

print("Loading Metadata...")

metadata = [

    ("employee", "employee.csv", 100, "Success", "2026-06-30"),
    ("department", "department.csv", 50, "Success", "2026-06-30"),
    ("customer", "customer.csv", 250, "Success", "2026-06-30"),
    ("sales", "sales.csv", 1000, "Success", "2026-06-30"),
    ("product", "product.csv", 150, "Success", "2026-06-30")

]

columns = [
    "TableName",
    "SourceFile",
    "RowsLoaded",
    "Status",
    "LoadDate"
]

df = spark.createDataFrame(metadata, columns)

df.write.mode("overwrite").saveAsTable("default.metadata")

print("Metadata Loaded Successfully.")

spark.sql("SELECT * FROM default.metadata").show()
```
